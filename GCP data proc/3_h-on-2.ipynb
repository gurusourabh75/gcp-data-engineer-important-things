```python
# create data frame 

emp=spark.createDataFrame(data=emp_data,schema=emp_schema)
```

```python
emp.printSchema()
```

```python
# antother way to check the schema 

emp.schema
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType
```

```python
# for columns and expression
# to import clumns and expression functions
from pyspark.sql.functions import col, expr 

col("name")
```

```python
# select column 
# select employ id, name , age, salary from emp 

emp_filtered= emp.select(col("employee_id"), expr("name"),emp.age,emp.salary)
```

```python
emp_filtered.show()
```

```python
# Using expr for select
# select employee_id as emp_id, name, cast (age as int) as age, salary from emp_filtered

emp_casted = emp_filtered.select(expr("employee_id as emp_id"),emp.name,expr("cast(age as int) as age"), emp.salary)
```

```python
# to show the data frame casted 
emp_casted.show()
```

```python
emp_casted.printSchema()
```

```python
# another mathod 

emp_casted_1=emp_filtered.selectExpr("employee_id as emp_id", "name", "cast(age as int) as age", "salary")
```

```python
emp_casted_1.show()
```

```python
emp_casted_1.printSchema()
```

```python
# Filter emp based on Age > 30
# select emp_id, name, age, salary from emp_casted where age â€º 30

emp_final = emp_casted.select("emp_id", "name", "salary").where("age > 30")

                              
```

```python
emp_final.show()
```

```python
# write the data back to csv ( action)

emp_final.write.csv("output/emp_final.csv", header=True)    
```
