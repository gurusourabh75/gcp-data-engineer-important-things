```python
# columns and expression 

from pyspark.sql.functions import col, expr

emp_df["salary"]
```

```python
# select columns
# select employe id , name , age salary from emp 

emp_filtered= emp_df.select(col("employee_id"),col("name"),col("age"),col("salary"))
emp_filtered.show()
```

```python
# Using expr for select
# select employee_id as emp_id, name, cast(age as int) as age, salary from emp_filtered

emp_casted=emp_filtered.select(expr("employee_id as emp_id"),emp_df.name,expr("cast(age as int)as age"),emp_df.salary)
emp_casted.show()
```

```python
# to change the column header with simple " selectExpr " and  define other 

emp_casted_1=emp_filtered.selectExpr("employee_id as emp_id","name", "cast(age as int) as age","salary")
emp_casted_1.show()
```

```python
# select  statement
#select emp_id,name age, salary where age > 30 

emp_final=emp_casted_1.select("emp_id","name","age","salary").where("age > 30 ")
emp_final.show()
```

```python
# write data back to source 

emp_final.write.format("csv")
```
